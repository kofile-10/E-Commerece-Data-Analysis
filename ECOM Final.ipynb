# *Training and Testing Data*

Splitting data into training and testing sets. 

X = Numerical Features

Y = Target Features/Variable

```python
y = ecom['Yearly Amount Spent']
```

```python
x = ecom[['Avg. Session Length', 'Time on App','Time on Website', 'Length of Membership']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=101)
```

## Training the Model

Model Training

```python
# Linear Regression Import
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

Print out the coefficients of the model

```python
# The coefficients
print('Coefficients: \n', lm.coef_)
```

## Predicting Test Data

```python
predictions = lm.predict(x_test)
```

```python
plt.scatter(y_test,predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

## Evaluating the Model

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

## Residuals

```python
sns.distplot((y_test-predictions),bins=50);
```

## Conclusion

```python
coeffecients = pd.DataFrame(lm.coef_,x.columns)
coeffecients.columns = ['Coeffecient']
coeffecients
```

A 1 unit increase in Avg. Session Length is associated with an increase of 25.98 total dollars spent.

A 1 unit increase in Time on App is associated with an increase of 38.59 total dollars spent.

A 1 unit increase in Time on Website is associated with an increase of 0.19 total dollars spent.

A 1 unit increase in Length of Membership is associated with an increase of 61.28 total dollars spent.


At face value, it would be more beneficial to focus on the App versus the Website.
